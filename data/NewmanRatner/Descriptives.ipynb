```python
import json
import os
```

```python
ADS = json.load(open("ADS.json", "r"))
```

```python
CDS = json.load(open("CDS.json", "r"))
```

```python
print(len(ADS))
print(len(CDS))
```

```python
def describe(dataset):
    descriptives = {"vocab": {},
                   "len_s": [],
                   "len_words": []}
    for utt in dataset:
        sent = "_".join(dataset[utt]['sentence'])
        clean_sent = sent.replace("_", " ")
        words = clean_sent.split()
        print(words)
        descriptives["len_words"].append(len(words))
        for word in words:
            if word in descriptives["vocab"]:
                descriptives["vocab"][word] += 1
            else:
                descriptives["vocab"][word] = 1
        descriptives["len_s"].append(dataset[utt]['end'] - dataset[utt]['start'])
    return descriptives
```

```python
# only select utterances that were actually included
ADS_clean = {}
CDS_clean = {}

# load the splits
ADS_splits = json.load(open("ADS_splits.json", "r"))
CDS_splits = json.load(open("CDS_splits.json", "r"))

splits = ["test", "val", "train"]
for split in splits:
    for utt in CDS_splits[split]:
        CDS_clean[utt] = CDS[utt]

for split in splits:
    for utt in ADS_splits[split]:
        ADS_clean[utt] = ADS[utt]

print(len(ADS_clean))
print(len(CDS_clean))
```

```python
descriptives_ADS = describe(ADS_clean)
descriptives_CDS = describe(CDS_clean)
```

```python
import numpy as np

def type_token(countdict):
    counts = [countdict[k] for k in countdict]
    tokens = np.sum(counts)
    types = len(countdict)
    return types/tokens

def print_descriptives(descriptives):
    print("Vocabulary size\t{}".format(len(descriptives["vocab"])))
    print("Total tokens\t{}".format(np.sum(descriptives["vocab"][k] for k in descriptives["vocab"])))
    print("Mean words per utterance\t{}".format(np.mean(descriptives["len_words"])))
    print("Mean length utterance in seconds\t{}".format(np.mean(descriptives["len_s"])))
    print("Words per sec\t{}".format(np.mean(descriptives["len_words"])/np.mean(descriptives["len_s"])))
    print("Type/token ratio\t{}".format(type_token(descriptives["vocab"])))
```

```python
print("ADS descriptives")
print_descriptives(descriptives_ADS)
print("CDS descriptives")
print_descriptives(descriptives_CDS)
```
