```python
import json
import matplotlib.pyplot as plt
import os
import ast
import numpy as np
```

```python
SEEDS = [123, 234, 345]
```

```python
resultfolders = ["../ads/results/",
              "../cds/results/",
              "../ads_synth/results/",
              "../cds_synth/results/"]
resultfiles = []
for folder in resultfolders:
    resultfiles.extend([folder + file for file in os.listdir(folder)])

retrieval_scores = {"ads": {}, "cds": {}, "ads_synth": {}, "cds_synth": {}}
for file in resultfiles:
    with open(file, "r") as f:
        if "/ads_synth/" in file:
            register = "ads_synth"
        elif "/cds_synth/" in file:
            register = "cds_synth"
        elif "/ads/" in file:
            register = "ads"
        elif "/cds/" in file:
            register = "cds"
        seed = int(file[-8:-5])
        text = f.readlines()
        results = [ast.literal_eval(line) for line in text]
        retrieval_scores[register][seed] = {'recall@1':[],
                                 'recall@5':[],
                                 'recall@10':[],
                                 'medr':[]}
        for r in results:
            retrieval_scores[register][seed]['recall@1'].append(r['recall'][1])
            retrieval_scores[register][seed]['recall@5'].append(r['recall'][5])
            retrieval_scores[register][seed]['recall@10'].append(r['recall'][10])
            retrieval_scores[register][seed]['medr'].append(r['medr'])
            
print(retrieval_scores)
```

```python
# crossvalidation plots
cross_scores = {}
cross_results = os.listdir("../crossval_results/")
for result in cross_results:
    with open ("../crossval_results/{}".format(result), "r") as f:
        data = json.load(f)
    cross_scores[result] = {"ADS": {}, "CDS": {}, "ADS_synth": {}, "CDS_synth": {}}
    for register in data.keys():
        for seed in data[register]:
            cross_scores[result][register][seed] = {'recall@1':[],
                                 'recall@5':[],
                                 'recall@10':[],
                                 'medr':[]}
            for e in range(0, 51):
                cross_scores[result][register][seed]['recall@1'].append(data[register][seed][str(e)]['recall']["1"])
                cross_scores[result][register][seed]['recall@5'].append(data[register][seed][str(e)]['recall']["5"])
                cross_scores[result][register][seed]['recall@10'].append(data[register][seed][str(e)]['recall']["10"])
                cross_scores[result][register][seed]['medr'].append(data[register][seed][str(e)]['medr'])
```

```python
for metric in ['recall@1', 'recall@5', 'recall@10','medr']:
    print(metric)
    for seed in SEEDS:
        for model in ["ADS", "CDS"]:
            if model == "ADS": 
                colour = "steelblue"
            elif model == "CDS":
                colour = "red"
            try:
                plt.plot(cross_scores["CDS.json"][model][str(seed)][metric], color=colour, linewidth = 0.7)
            except:
                pass
    plt.grid()
    plt.show()
```
